```python
# ATM Machine Simulation
def atm_machine():
    pin = "1234"        
    balance = 10000    
    transaction_history = []   # to store past transactions
    chances = 3                # max attempts for PIN
    
    print("===== Welcome to ATM Machine =====") 
    
    # PIN Verification with 3 attempts
    while chances > 0:
        entered_pin = input("Enter your 4-digit PIN: ")
        if entered_pin == pin:
            print("\n Access Granted. Welcome!\n")
            break
        else:
            chances -= 1
            print(f" Incorrect PIN! Attempts left: {chances}")
            if chances == 0:
                print(" Card Blocked! Too many wrong attempts.")
                return
    
    # Keep showing menu until Exit
    while True:
        print("\n=========== ATM Menu ===========")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. View Transaction History")
        print("5. Change PIN")
        print("6. Exit")
        print("================================")
        
        choice = input("Enter your choice: ")
        
        # 1. Balance Check
        if choice == "1":
            print(f"Your current balance is: Rs. {balance}")
        
        # 2. Deposit Money
        elif choice == "2":
            amount = int(input("Enter amount to deposit: Rs"))
            if amount > 0:
                balance += amount
                transaction_history.append(f"Deposited Rs. {amount}")
                print(f"Deposit successful! New balance: Rs. {balance}")
            else:
                print("Invalid amount entered.")
        
        # 3. Withdraw Money
        elif choice == "3":
            amount = int(input("Enter amount to withdraw: Rs."))
            if amount <= 0:
                print("Enter a valid amount.")
            elif amount <= balance:
                balance -= amount
                transaction_history.append(f"Withdrew Rs. {amount}")
                print(f"Withdrawal successful! Remaining balance: Rs. {balance}")
            else:
                print("Insufficient balance!")
        
        # 4. View Transaction History
        elif choice == "4":
            print("\n Transaction History:")
            if len(transaction_history) == 0:
                print("No transactions yet.")
            else:
                for t in transaction_history:
                    print(t)   
        
        # 5. Change PIN
        elif choice == "5":
            old_pin = input("Enter your old PIN: ")
            if old_pin == pin:
                new_pin = input("Enter new 4-digit PIN: ")
                confirm_pin = input("Confirm new PIN: ")
                if new_pin == confirm_pin and len(new_pin) == 4 and new_pin.isdigit():
                    pin = new_pin
                    print(" PIN changed successfully!")
                else:
                    print("PIN change failed! Please try again.")
            else:
                print("Wrong old PIN!")

        # 6. Exit
        elif choice == "6":
            print(" Thank you for using ATM. Goodbye!")
            break   
        
        else:
            print(" Invalid choice! Please try again.")

atm_machine()
```
